```python
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Directory containing the CSV files
csv_directory = "D:\MTdataset\DL_dataset\Python_DL\Final_current_vibration_data"
```

```python
# Function to read CSV files and generate summary
def generate_summary(csv_files):
    summary = {}
    for csv_file in csv_files:
        df = pd.read_csv(os.path.join(csv_directory, csv_file))
        summary[csv_file] = {
            'Columns': df.columns.tolist(),
            'Number of Rows': len(df)
        }
    return summary
```

```python
# Function to display the summary in a table format
def display_summary_table(summary, title):
    df_summary = pd.DataFrame(summary).T
    print(f"\n{title}")
    print(df_summary)
    df_summary.plot(kind='barh', figsize=(10, 6))
    plt.title(title)
    plt.xlabel('Count')
    plt.show()
```

```python
# Function to perform data exploration
def explore_data(csv_files):
    for csv_file in csv_files:
        df = pd.read_csv(os.path.join(csv_directory, csv_file))
        print(f"\nExploring {csv_file}:")

        # Checking for missing values
        missing_values = df.isnull().sum()
        print("Missing Values:\n", missing_values)

        # Checking for string values in numeric columns
        for column in df.columns:
            if df[column].dtype == object:
                print(f"String values found in column {column}")

        # Display basic statistics
        print("Basic Statistics:\n", df.describe())
```

```python
# List of CSV files
csv_files = [file for file in os.listdir(csv_directory) if file.endswith('.csv')]
```

```python
# Separate the CSV files into 'current' and 'vibration'
current_csv_files = [file for file in csv_files if 'current' in file]
vibration_csv_files = [file for file in csv_files if 'vibration' in file]
```

```python
# Generate summaries
current_summary = generate_summary(current_csv_files)
vibration_summary = generate_summary(vibration_csv_files)
```

```python
# Display summaries
display_summary_table(current_summary, 'Current CSV Files Summary')
display_summary_table(vibration_summary, 'Vibration CSV Files Summary')
```

```python
# Perform data exploration
explore_data(current_csv_files)
explore_data(vibration_csv_files)
```
